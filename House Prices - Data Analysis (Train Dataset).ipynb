### Data Analysis Phase 

the main for this project is to predict the house price based on variouse freatures. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns  

pd.pandas.set_option("Display.max_columns", None)
```

```python
dataset=pd.read_csv("train.csv")
print(dataset.shape)
```

```python
dataset.head()
```

#### Missing Values 

```python
## Find How much missing values are present in data
feature_with_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1]
for feature in feature_with_nan:
    print(feature, np.round(dataset[feature].isnull().mean(), 4), "% is missing values")
```

```python
## Find the relationship between missing values and SalePrice
for feature in feature_with_nan:
    data=dataset.copy()
    ## 1 means null value is persent ontherwise o(not null value)
    data[feature]= np.where(data[feature].isnull(), 1, 0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
print("total couunt of Id {}".format(len(dataset.Id)))
```

###  Numerical Variable 

```python
## find the list of numerical variable

numerical_feature=[feature for feature in dataset.columns if dataset[feature].dtype != 'O']
print('number of numerical variable', len(numerical_feature))
dataset[numerical_feature].head()
```

####  trying to find out Temporal variable. (Eg:- DateTime feature(year or month or day))

```python
year_feature=[feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature ]
 
year_feature
```

```python
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
## Analyze the year_feature with YrSold data 

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Sales Price')
plt.title(feature)
plt.show()
```

```python
## now try to compare All year feature with sales price

for feature in year_feature:
    if feature != 'YrSold':
        data=dataset.copy()
        data[feature]= data['YrSold']-data[feature]
        
        
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel("SalePrice")
        plt.title('year feature')
        plt.show()
    
```

```python
## In numerical variable their are 2 type
#### 1. continous variable 2. Discrete Variable

#discrete variable 
discrete_feature=[feature for feature in numerical_feature if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print('total discrete features is {}'.format(len(discrete_feature)))
```

```python
## find the relationship between all the discrete variable and sales price

for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
## Continous Variable 

continous_feature=[feature for feature in numerical_feature if feature not in discrete_feature+year_feature+['Id']]
print("total continous variable is {}".format(len(continous_feature)))
```

```python
for feature in continous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('sale price')
    plt.title(feature)
    plt.show()
```

```python
## We will using logarithmin trasformation 

for feature in continous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
        
```

## Outlier

```python
for feature in continous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```
